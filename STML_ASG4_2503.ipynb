## Question 1:

https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification?select=train.csvLinks to an external site.

From the above data:

1.Read the data with pandas and find features and target variables 2.Normalize the data with min-max scaling 3.Split the data into train and test.

```python
import pandas as pd
d=pd.read_csv("train.csv")
print(d)
```

```python
y=d['price_range']
print(y)
```

```python
x=d.drop('price_range',axis=1)
print(x)
```

```python
dd=(x-x.min())/(x.max()-x.min())
print(dd)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dd,y,test_size=100)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
target = d['price_range']
print(target)
```

```python
feature = d.drop('price_range', axis=1)
print(feature)
```

```python
d.isnull()
```

```python
normalised_data = (feature - feature.min()) / (feature.max() - feature.min())
print(normalised_data)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(normalised_data,target,test_size=100)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
