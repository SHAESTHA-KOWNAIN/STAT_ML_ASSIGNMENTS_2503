### ***Find out the maximum likelihood parameters values for california_housing [‘latitude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, each column individually.***

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
print(df)

import pandas as pd
import numpy as np

#latitude mean and variance
latitude_mean = df['latitude'].mean()
latitude_var = df['latitude'].var()

#total rooms mean and variance
total_rooms_mean = df['total_rooms'].mean()
total_rooms_var = df['total_rooms'].var()

# House Median Age
house_median_age_mean = df['housing_median_age'].mean()
house_median_age_var = df['housing_median_age'].var()

# Total Bedrooms
total_bedrooms_mean = df['total_bedrooms'].mean()
total_bedrooms_var = df['total_bedrooms'].var()

print(f'Latitude: Mean = {latitude_mean}, Variance = {latitude_var}')
print(f'Total Rooms: Mean = {total_rooms_mean}, Variance = {total_rooms_var}')
print(f'House Median Age: Mean = {house_median_age_mean}, Variance = {house_median_age_var}')
print(f'Total Bedrooms: Mean = {total_bedrooms_mean}, Variance = {total_bedrooms_var}')
```

### ***Find log-likelihood values for 50, 75, and 80. For California_housing [‘latitude’] and find the Max likelihood values.***

```python
import numpy as np
latitude_var = df['latitude'].var()
# Given values
x = [50, 75, 80]

# Calculate log-likelihood for each value
log_likelihoods = []
for i in x:
    log_likelihood = np.sum(
        -0.5 * np.log(2 * np.pi * latitude_var)
        - 0.5 * ((i - df['latitude']) ** 2) / latitude_var
    )
    log_likelihoods.append(log_likelihood)
    print(f'Log-likelihood for {i}: {log_likelihood}')

# Find the value with the maximum log-likelihood
max_likelihood_value = x[np.argmax(log_likelihoods)]
print(f'\nValue with maximum log-likelihood: {max_likelihood_value}')
```
